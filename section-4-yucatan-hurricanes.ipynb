## Week 4: Yucatan Peninsula Hurricanes

### 1. Archive exploration

This data set contains hurricane data and their category at land fall, as well as reconstructed maximum damage across the Yucatan Peninsula. They used computer models to recontruct impacts of 105 storms.

Boose, E. and D. Foster. 2023. Ecological Impacts of Hurricanes Across the Yucatan Peninsula 1851-2000 ver 23. Environmental Data Initiative. https://doi.org/10.6073/pasta/f219113373913f2daf421732e28d3c38 (Accessed 2025-10-24).

### 2. Data loading and exploration

```python
import pandas as pd

# Read in data
URL = 'https://pasta.lternet.edu/package/data/eml/knb-lter-hfr/71/23/ab0fe2bf4f3ad850371ccb9c69d78469'
df = pd.read_csv(URL)

# Check first few rows 
df.head()
```

```python
# Filter to Saffir-Simpson category 5
cat5 = df.loc[df.ss == 5]

# Check number of rows
cat5.shape

# Check type of date
cat5.dtypes
```

```python
cat5['start.date'] = pd.to_datetime(cat5['start.date'])

cat5['end.date'] = pd.to_datetime(cat5['end.date'])

cat5['duration'] = cat5['end.date'] - cat5['start.date']

cat5.head()
```

### 5. Visualize categories across time

```python
import matplotlib.pyplot as plt

# Update start date to datetime
df['start.date'] = pd.to_datetime(df['start.date'])

# Create plot
fig, ax = plt.subplots(figsize = (10,5))

ax.set_yticks([1, 2, 3, 4, 5])
ax.scatter(df['start.date'], df['ss'], color = 'purple')
ax.set(title = "Saffir-Simpson categories across time",
      xlabel = 'Start date',
      ylabel = 'Category')

plt.show()
```

### 6. Collect code

```python
import pandas as pd
import matplotlib.pyplot as plt

# Read in data
URL = 'https://pasta.lternet.edu/package/data/eml/knb-lter-hfr/71/23/ab0fe2bf4f3ad850371ccb9c69d78469'
df = pd.read_csv(URL)

# Update start date to datetime
df['start.date'] = pd.to_datetime(df['start.date'])

# Create plot
fig, ax = plt.subplots()

ax.set_yticks([1, 2, 3, 4, 5])
ax.scatter(df['start.date'], df['ss'], color = 'purple')
ax.set(title = "Saffir-Simpson categories across time",
      xlabel = 'Year',
      ylabel = 'Category')
```
